```python
# ======================================================
# GOLD PRICE ANALYSIS + MACHINE LEARNING VISUALIZATION
# Includes Line charts, Moving Averages, Volatility, Candlestick, Interactive Plotly, ML Models
# Author: Elsie Owusu Ansah
# Date: 2026-03-13
# ======================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import mplfinance as mpf

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, roc_curve, auc

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

# -------------------------------
# 1 Create output folder
# -------------------------------
os.makedirs("visuals", exist_ok=True)

# -------------------------------
# 2 Load dataset
# -------------------------------
df = pd.read_csv("GoldUSD.csv")

df.columns = df.columns.str.strip().str.lower()
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")

# -------------------------------
# 3 Create year ticks
# -------------------------------
year_ticks = pd.to_datetime([
"2000-01-01","2004-01-01","2008-01-01",
"2012-01-01","2016-01-01","2020-01-01",
"2024-01-01","2028-01-01"
])

# ===============================
# GOLD CLOSING PRICE TREND
# ===============================
plt.figure(figsize=(14,7))

plt.plot(df["date"], df["close"], color="gold", linewidth=2)

plt.title("Gold Closing Price Trend")
plt.xlabel("Date")
plt.ylabel("Price")

plt.xticks(year_ticks, year_ticks.year)

plt.grid(alpha=0.3)
plt.tight_layout()

plt.savefig("visuals/gold_closing_trend.png", dpi=300)
plt.show()

# ===============================
# MOVING AVERAGES
# ===============================
df["MA50"] = df["close"].rolling(50).mean()
df["MA200"] = df["close"].rolling(200).mean()

plt.figure(figsize=(14,7))

plt.plot(df["date"], df["close"], label="Close")
plt.plot(df["date"], df["MA50"], label="50 Day MA")
plt.plot(df["date"], df["MA200"], label="200 Day MA")

plt.title("Gold Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price")

plt.xticks(year_ticks, year_ticks.year)

plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.savefig("visuals/moving_average.png", dpi=300)
plt.show()

# ===============================
# VOLATILITY
# ===============================
returns = df["close"].pct_change()
volatility = returns.rolling(30).std()

plt.figure(figsize=(14,7))

plt.plot(df["date"], volatility, color="purple")

plt.title("Gold Price Volatility (30 Day)")
plt.xlabel("Date")
plt.ylabel("Volatility")

plt.xticks(year_ticks, year_ticks.year)

plt.grid(alpha=0.3)
plt.tight_layout()

plt.savefig("visuals/volatility.png", dpi=300)
plt.show()

# ===============================
# TRADING VOLUME
# ===============================
plt.figure(figsize=(14,7))

plt.bar(df["date"], df["volume"], color="red")

plt.title("Gold Trading Volume")
plt.xlabel("Date")
plt.ylabel("Volume")

plt.xticks(year_ticks, year_ticks.year)

plt.tight_layout()
plt.savefig("visuals/trading_volume.png", dpi=300)
plt.show()

# ===============================
# OHLC MOVEMENT
# ===============================
plt.figure(figsize=(14,7))

plt.plot(df["date"], df["open"], label="Open")
plt.plot(df["date"], df["high"], label="High")
plt.plot(df["date"], df["low"], label="Low")
plt.plot(df["date"], df["close"], label="Close")

plt.title("Gold Price Movement (OHLC)")
plt.xlabel("Date")
plt.ylabel("Price")

plt.xticks(year_ticks, year_ticks.year)

plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.savefig("visuals/ohlc.png", dpi=300)
plt.show()

# ===============================
# CANDLESTICK CHART
# ===============================
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

# Clean column names
df.columns = df.columns.str.strip().str.capitalize()

# Convert Date column
df["Date"] = pd.to_datetime(df["Date"])

# Sort values
df = df.sort_values("Date")

plt.figure(figsize=(14,7))

for i in range(len(df)):

    open_price = df["Open"].iloc[i]
    close_price = df["Close"].iloc[i]
    high_price = df["High"].iloc[i]
    low_price = df["Low"].iloc[i]
    date = df["Date"].iloc[i]

    color = "green" if close_price > open_price else "red"

    # wick
    plt.plot([date, date], [low_price, high_price], color="black")

    # candle body
    plt.plot([date, date], [open_price, close_price], color=color, linewidth=5)

# --------- FORCE DATE RANGE ----------
start = pd.to_datetime("2000-01-01")
end = pd.to_datetime("2028-01-01")

plt.xlim(start, end)

ticks = pd.to_datetime([
"2000-01-01",
"2004-01-01",
"2008-01-01",
"2012-01-01",
"2016-01-01",
"2020-01-01",
"2024-01-01",
"2028-01-01"
])

plt.xticks(ticks, [2000,2004,2008,2012,2016,2020,2024,2028])

# -------------------------------------

plt.title("Gold Candlestick Chart", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Price")

plt.grid(alpha=0.3)
plt.xticks(rotation=45)

plt.tight_layout()

plt.savefig("visuals/gold_candlestick.png", dpi=300)

plt.show()
# ===============================
# CORRELATION HEATMAP
# ===============================
plt.figure(figsize=(10,6))

corr = df[["open","high","low","close","volume"]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Between Gold Price Variables")

plt.tight_layout()
plt.savefig("visuals/correlation_heatmap.png", dpi=300)
plt.show()

# ======================================================
# MACHINE LEARNING SECTION
# ======================================================

# Predict if next day price goes up
df["target"] = (df["close"].shift(-1) > df["close"]).astype(int)

features = df[["open","high","low","close","volume"]]
target = df["target"]

features = features.dropna()
target = target.loc[features.index]

X_train, X_test, y_train, y_test = train_test_split(
features, target, test_size=0.2, random_state=42
)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

models = {
"Logistic Regression": LogisticRegression(),
"Decision Tree": DecisionTreeClassifier(),
"Random Forest": RandomForestClassifier(),
"SVM": SVC(probability=True),
"KNN": KNeighborsClassifier(),
"Naive Bayes": GaussianNB()
}

# ===============================
# CONFUSION MATRICES + ROC
# ===============================

for name, model in models.items():

    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    probs = model.predict_proba(X_test)[:,1]

    cm = confusion_matrix(y_test, preds)

    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")

    plt.title(f"Confusion Matrix - {name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")

    plt.tight_layout()
    plt.savefig(f"visuals/confusion_{name.replace(' ','_')}.png", dpi=300)
    plt.show()

    fpr, tpr, _ = roc_curve(y_test, probs)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(7,6))

    plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
    plt.plot([0,1],[0,1],'--')

    plt.title(f"ROC Curve - {name}")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")

    plt.legend()

    plt.tight_layout()
    plt.savefig(f"visuals/roc_{name.replace(' ','_')}.png", dpi=300)
    plt.show()

print("All charts successfully generated in visuals folder.")
```

```python
# -----------------------------------------
# GOLD MARKET VOLATILITY DASHBOARD
# -----------------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------
# 1 Load dataset
# -------------------------

df = pd.read_csv("GoldUSD.csv")

df.columns = df.columns.str.lower()
df["date"] = pd.to_datetime(df["date"])

df = df.sort_values("date")

# -------------------------
# 2 Feature Engineering
# -------------------------

# Daily returns
df["returns"] = df["close"].pct_change()

# 30-day volatility index
df["volatility"] = df["returns"].rolling(30).std()

# Rolling means
df["mean_price"] = df["close"].rolling(180).mean()
df["mean_volume"] = df["volume"].rolling(180).mean()

# Simulated macroeconomic indicators
np.random.seed(42)
df["inflation_proxy"] = df["close"].pct_change().rolling(60).mean()
df["dollar_strength"] = -df["returns"].rolling(60).mean()
df["rate_pressure"] = df["volatility"].rolling(60).mean()

# -------------------------
# 3 Create Dashboard
# -------------------------

fig, axes = plt.subplots(3,2, figsize=(18,12))

# -------------------------
# Gold Price Trend
# -------------------------

axes[0,0].plot(df["date"], df["close"], label="Gold Price", color="gold")
axes[0,0].plot(df["date"], df["mean_price"], label="Mean Price (180d)", color="black")

axes[0,0].set_title("Gold Price Trend")
axes[0,0].legend()
axes[0,0].grid(True)

# -------------------------
# Volatility Index
# -------------------------

axes[0,1].plot(df["date"], df["volatility"], color="red")

axes[0,1].set_title("Gold Volatility Index (30-Day)")
axes[0,1].grid(True)

# -------------------------
# Trading Volume
# -------------------------

axes[1,0].plot(df["date"], df["volume"], color="blue", alpha=0.6)
axes[1,0].plot(df["date"], df["mean_volume"], color="black", label="Mean Volume")

axes[1,0].set_title("Gold Trading Volume")
axes[1,0].legend()
axes[1,0].grid(True)

# -------------------------
# Mean Price vs Volume
# -------------------------

axes[1,1].scatter(df["mean_price"], df["mean_volume"], alpha=0.5)

axes[1,1].set_title("Mean Price vs Mean Volume")
axes[1,1].set_xlabel("Mean Price")
axes[1,1].set_ylabel("Mean Volume")

# -------------------------
# Macroeconomic Signals
# -------------------------

axes[2,0].plot(df["date"], df["inflation_proxy"], label="Inflation Signal")
axes[2,0].plot(df["date"], df["dollar_strength"], label="Dollar Strength")
axes[2,0].plot(df["date"], df["rate_pressure"], label="Rate Pressure")

axes[2,0].set_title("Macroeconomic Signals")
axes[2,0].legend()
axes[2,0].grid(True)

# -------------------------
# Correlation Heatmap
# -------------------------

corr = df[["close","volume","volatility",
           "inflation_proxy","dollar_strength","rate_pressure"]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", ax=axes[2,1])

axes[2,1].set_title("Market Signal Correlation")

# -------------------------
# Layout
# -------------------------

plt.suptitle("Gold Market Volatility Dashboard", fontsize=18)

plt.tight_layout()

plt.savefig("gold_market_dashboard.png", dpi=300)

plt.show()
```

```python
# ---------------------------
# Hedge Fund Style Price + Volatility Regime Chart
# ---------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Load data
df = pd.read_csv("GoldUSD.csv")
df.columns = df.columns.str.lower()
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")

# Feature Engineering
df["returns"] = df["close"].pct_change()
df["volatility"] = df["returns"].rolling(30).std()
df["mean_price"] = df["close"].rolling(180).mean()

# ---------------------------
# Dual-axis plot
# ---------------------------
fig, ax1 = plt.subplots(figsize=(16,7))

# Gold price trend
ax1.plot(df["date"], df["close"], color="gold", label="Gold Price ($/oz)", linewidth=2)
ax1.plot(df["date"], df["mean_price"], color="black", linestyle="--", label="180d Rolling Mean")

ax1.set_xlabel("Year")
ax1.set_ylabel("Price (USD/oz)")
ax1.tick_params(axis="y", labelcolor="black")
ax1.grid(True, alpha=0.3)

# Highlight crisis periods
crises = [
    ("2008-09-01", "2009-06-01", "Financial Crisis"),
    ("2020-02-01", "2020-08-01", "COVID-19 Crash"),
    ("2021-06-01", "2022-06-01", "Inflation Spike")
]
for start, end, label in crises:
    ax1.axvspan(pd.to_datetime(start), pd.to_datetime(end), color="red", alpha=0.2)
    ax1.text(pd.to_datetime(start), df["close"].max()*0.95, label, color="red", fontsize=10, rotation=90)

# Volatility on secondary axis
ax2 = ax1.twinx()
ax2.plot(df["date"], df["volatility"], color="red", label="30-Day Volatility", linewidth=1.5, alpha=0.7)
ax2.set_ylabel("Volatility (30d Std. Dev.)", color="red")
ax2.tick_params(axis="y", labelcolor="red")

# Combine legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc="upper left")

# X-axis: year ticks
ax1.set_xticks(pd.date_range(start="2000-01-01", end="2028-01-01", freq="4YS"))
ax1.set_xticklabels([str(year) for year in range(2000,2029,4)])

plt.title("Gold Price and Volatility Regimes (Hedge Fund Style)")
plt.tight_layout()
plt.savefig("gold_price_volatility_regimes.png", dpi=300)
plt.show()
```

```python
# -----------------------------------------
# MONTHLY GOLD MARKET DASHBOARD - 2008
# -----------------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------
# 1. Load dataset
# -------------------------
df = pd.read_csv("GoldUSD.csv")
df.columns = df.columns.str.lower()
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")

# Filter for 2008
df_2008 = df[(df["date"].dt.year == 2008)].copy()

# -------------------------
# 2. Aggregate Monthly Metrics
# -------------------------
df_2008.set_index("date", inplace=True)
monthly = df_2008.resample("M").agg({
    "open":"first",
    "high":"max",
    "low":"min",
    "close":"mean",
    "volume":"sum"
})

# Calculate Returns, Volatility, Drawdown
monthly["returns"] = monthly["close"].pct_change()
monthly["volatility"] = monthly["returns"].rolling(3).std()  # quarterly rolling volatility
monthly["cumulative"] = (1 + monthly["returns"]).cumprod()
monthly["rolling_max"] = monthly["cumulative"].cummax()
monthly["drawdown"] = (monthly["cumulative"] - monthly["rolling_max"]) / monthly["rolling_max"]

# 3-Month Moving Average
monthly["sma_3"] = monthly["close"].rolling(3).mean()

# -------------------------
# 3. Plot Dashboard
# -------------------------
fig, axes = plt.subplots(3,2, figsize=(16,12))

# Gold Price Trend
axes[0,0].plot(monthly.index, monthly["close"], label="Monthly Avg Close", color="gold", linewidth=2)
axes[0,0].plot(monthly.index, monthly["sma_3"], label="3-Month SMA", color="black", linestyle="--")
axes[0,0].set_title("Gold Price Trend - 2008")
axes[0,0].set_xlabel("Month")
axes[0,0].set_ylabel("Price (USD)")
axes[0,0].legend()
axes[0,0].grid(True)

# Volatility
axes[0,1].bar(monthly.index, monthly["volatility"], color="red", alpha=0.6)
axes[0,1].set_title("Monthly Volatility (Returns Std)")
axes[0,1].set_xlabel("Month")
axes[0,1].set_ylabel("Volatility")
axes[0,1].grid(True)

# Trading Volume
axes[1,0].bar(monthly.index, monthly["volume"], color="purple", alpha=0.6)
axes[1,0].set_title("Monthly Trading Volume")
axes[1,0].set_xlabel("Month")
axes[1,0].set_ylabel("Volume (oz)")
axes[1,0].grid(True)

# Drawdown
axes[1,1].plot(monthly.index, monthly["drawdown"], color="darkred", linewidth=2)
axes[1,1].set_title("Drawdown Risk")
axes[1,1].set_xlabel("Month")
axes[1,1].set_ylabel("Drawdown (%)")
axes[1,1].grid(True)

# Open vs Close Range
axes[2,0].bar(monthly.index, monthly["high"]-monthly["low"], bottom=monthly["low"], color="pink", alpha=0.7)
axes[2,0].set_title("Monthly High-Low Range")
axes[2,0].set_xlabel("Month")
axes[2,0].set_ylabel("Price Range (USD)")
axes[2,0].grid(True)

# Correlation Heatmap
corr = monthly[["open","high","low","close","volume","volatility"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", ax=axes[2,1])
axes[2,1].set_title("Correlation Between Monthly Metrics")

plt.suptitle("Monthly Gold Market Dashboard - 2008 Financial Crisis", fontsize=18)
plt.tight_layout()
plt.savefig("gold_dashboard_2008.png", dpi=300)
plt.show()
```

```python
# -----------------------------------------
# MONTHLY GOLD MARKET DASHBOARD - 2020 & 2022
# -----------------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------
# 1. Load dataset
# -------------------------
df = pd.read_csv("GoldUSD.csv")
df.columns = df.columns.str.lower()
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")

# Function to generate monthly dashboard
def gold_monthly_dashboard(df, year, save_path):
    df_year = df[df["date"].dt.year == year].copy()
    df_year.set_index("date", inplace=True)

    # Aggregate Monthly Metrics
    monthly = df_year.resample("M").agg({
        "open":"first",
        "high":"max",
        "low":"min",
        "close":"mean",
        "volume":"sum"
    })

    # Returns, Volatility, Drawdown
    monthly["returns"] = monthly["close"].pct_change()
    monthly["volatility"] = monthly["returns"].rolling(3).std()
    monthly["cumulative"] = (1 + monthly["returns"]).cumprod()
    monthly["rolling_max"] = monthly["cumulative"].cummax()
    monthly["drawdown"] = (monthly["cumulative"] - monthly["rolling_max"]) / monthly["rolling_max"]

    # 3-Month SMA
    monthly["sma_3"] = monthly["close"].rolling(3).mean()

    # -------------------------
    # 3. Plot Dashboard
    # -------------------------
    fig, axes = plt.subplots(3,2, figsize=(16,12))

    # Price Trend
    axes[0,0].plot(monthly.index, monthly["close"], label="Monthly Avg Close", color="gold", linewidth=2)
    axes[0,0].plot(monthly.index, monthly["sma_3"], label="3-Month SMA", color="black", linestyle="--")
    axes[0,0].set_title(f"Gold Price Trend - {year}")
    axes[0,0].set_xlabel("Month")
    axes[0,0].set_ylabel("Price (USD)")
    axes[0,0].legend()
    axes[0,0].grid(True)

    # Volatility
    axes[0,1].bar(monthly.index, monthly["volatility"], color="red", alpha=0.6)
    axes[0,1].set_title("Monthly Volatility (Returns Std)")
    axes[0,1].set_xlabel("Month")
    axes[0,1].set_ylabel("Volatility")
    axes[0,1].grid(True)

    # Trading Volume
    axes[1,0].bar(monthly.index, monthly["volume"], color="blue", alpha=0.6)
    axes[1,0].set_title("Monthly Trading Volume")
    axes[1,0].set_xlabel("Month")
    axes[1,0].set_ylabel("Volume (oz)")
    axes[1,0].grid(True)

    # Drawdown
    axes[1,1].plot(monthly.index, monthly["drawdown"], color="darkred", linewidth=2)
    axes[1,1].set_title("Drawdown Risk")
    axes[1,1].set_xlabel("Month")
    axes[1,1].set_ylabel("Drawdown (%)")
    axes[1,1].grid(True)

    # High-Low Range
    axes[2,0].bar(monthly.index, monthly["high"]-monthly["low"], bottom=monthly["low"], color="orange", alpha=0.7)
    axes[2,0].set_title("Monthly High-Low Range")
    axes[2,0].set_xlabel("Month")
    axes[2,0].set_ylabel("Price Range (USD)")
    axes[2,0].grid(True)

    # Correlation Heatmap
    corr = monthly[["open","high","low","close","volume","volatility"]].corr()
    sns.heatmap(corr, annot=True, cmap="coolwarm", ax=axes[2,1])
    axes[2,1].set_title("Correlation Between Monthly Metrics")

    plt.suptitle(f"Monthly Gold Market Dashboard - {year}", fontsize=18)
    plt.tight_layout()
    plt.savefig(save_path, dpi=300)
    plt.show()

# -------------------------
# Generate Dashboards
# -------------------------
gold_monthly_dashboard(df, 2020, "gold_dashboard_2020.png")
gold_monthly_dashboard(df, 2022, "gold_dashboard_2022.png")
```
